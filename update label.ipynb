```python
import time 

a = np.random.rand(1000000)
b = np.random.rand(1000000)

awal = time.time()
c = np.dot(a,b)
akhir = time.time()

print(c)
print("vectorization version "+ str(1000*(akhir-awal)) + "ms")

c = 0
awal = time.time()
for i in range(1000000):
    c+= a[i]*b[i]
akhir = time.time()

print(c)
print("for loop version "+ str(1000*(akhir-awal)) + "ms")
```
