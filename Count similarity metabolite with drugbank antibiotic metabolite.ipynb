```python
from rdkit import DataStructs
import rdkit as  rd
from rdkit import Chem
```

```python
hasil_similarity = pd.DataFrame()
temp = []


for j in range(14):
    for i in data["smiles"]:    
        ref = Chem.MolFromSmiles(df['smiles'][j])
        mol1 = Chem.MolFromSmiles(i)
        fp1 = Chem.RDKFingerprint(ref)
        fp2 = Chem.RDKFingerprint(mol1)
        Tan  = DataStructs.AsymmetricSimilarity(fp1,fp2)
        temp.append(Tan)
    hasil_similarity[df["Drug antibiotic"][j]] = temp
    temp = []
```

```python
hasil_similarity["Metabolite_Name"] = data["Metabolite_name"]
```
