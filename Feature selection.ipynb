```python
import pandas as pd
import numpy  as np
import gc
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import lazypredict
from lazypredict.Supervised import LazyClassifier
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_excel('data_diseases.xlsx')
X = df.iloc[:,0:465]
y = df.iloc[:,465:]
```

```python
feature_name = X.columns.tolist()
```

```python
def cor_selector(X, y):
    cor_list = []
    # calculate the correlation with y for each feature
    for i in X.columns.tolist():
        cor = np.corrcoef(X[i], y)[0, 1]
        cor_list.append(cor)
    # replace NaN with 0
    cor_list = [0 if np.isnan(i) else i for i in cor_list]
    # feature name
    cor_feature = X.iloc[:,np.argsort(np.abs(cor_list))[-100:]].columns.tolist()
    # feature selection? 0 for not select, 1 for select
    cor_support = [True if i in cor_feature else False for i in feature_name]
    return cor_support, cor_feature
```

```python
cor_support, cor_feature = cor_selector(X, y)
print(str(len(cor_feature)), 'selected features')
```

```python
# Selection fitur use Pearson Correlation
```

```python
X_new = X[cor_feature]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X_new, y, test_size=0.33)
```

```python
clf = LazyClassifier(verbose=0,ignore_warnings=True)
models, predictions = clf.fit(x_train, x_test, y_train, y_test)
models
```

```python
# fitur selection use chi-2
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.preprocessing import MinMaxScaler
X_norm = MinMaxScaler().fit_transform(X)
chi_selector = SelectKBest(chi2, k=200)
chi_selector.fit(X_norm, y)
```

```python
chi_support = chi_selector.get_support()
chi_feature = X.loc[:,chi_support].columns.tolist()
print(str(len(chi_feature)), 'selected features')
```

```python
X_new = X[chi_feature]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X_new, y, test_size=0.33)
```

```python
clf = LazyClassifier(verbose=0,ignore_warnings=True)
models, predictions = clf.fit(x_train, x_test, y_train, y_test)
models
```

```python
# fitur selection use Wrapper
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
rfe_selector = RFE(estimator=LogisticRegression(), n_features_to_select=100, step=10, verbose=5)
rfe_selector.fit(X_norm, y)
```

```python
rfe_support = rfe_selector.get_support()
rfe_feature = X.loc[:,rfe_support].columns.tolist()
print(str(len(rfe_feature)), 'selected features')
```

```python
X_new = X[rfe_feature]
x_train, x_test, y_train, y_test = train_test_split(X_new, y, test_size=0.33)
```

```python
clf = LazyClassifier(verbose=0,ignore_warnings=True)
models, predictions = clf.fit(x_train, x_test, y_train, y_test)
models
```
